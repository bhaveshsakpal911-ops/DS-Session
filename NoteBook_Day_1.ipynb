```python
def issue_book(student_name, books):
  print(f"\n{student_name} borrowed:")
  for book in books:
    print("-", book)

def calculate_fine(days_late):
  fine_per_day = 5
  return days_late * fine_per_day

def student_report(student_name, books, days_late):
  issue_book(student_name, books) # Corrected typo: issue_books -> issue_book
  fine = calculate_fine(days_late)
  print(f"Fine to be paid: ${fine}\n")

student_report("Rahul",["Python Basics", "Data Science 101"], 3)
student_report("Priya", ["Machine Learning"], 0)
```
